```python
# Exercise 1 - Time Series Forecasting

import numpy as np
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Dense, Flatten, SimpleRNN
from keras.optimizers import Adam

# Generating dataset
def generate_time_series(batch_size, n_steps):
    freq1, freq2, offsets1, offsets2 = np.random.rand(4, batch_size, 1)
    time = np.linspace(0, 1, n_steps)

    series = 0.5 * np.sin((time - offsets1) * (freq1 * 10 + 10))
    series += 0.2 * np.sin((time - offsets2) * (freq2 * 20 + 20))
    series += 0.1 * (np.random.rand(batch_size, n_steps) - 0.5)
    return series[..., np.newaxis].astype(np.float32)

data_size = 10000
n_steps = 50
np.random.seed(0)

series = generate_time_series(data_size, n_steps + 1)

# Train/Valid/Test split
X_train, y_train = series[:7000, :n_steps], series[:7000, -1]
X_valid, y_valid = series[7000:9000, :n_steps], series[7000:9000, -1]
X_test, y_test = series[9000:, :n_steps], series[9000:, -1]

no_training_epochs = 5

# Data examination
plt.figure()
plt.plot(X_train[0])
plt.plot(n_steps, y_train[0], 'rx')
plt.title("Sample Time Series and Target")
plt.xlabel("Time step")
plt.ylabel("Value")
plt.grid(True)
plt.show()

# Plot training history
def plot_function(history, title="Training History"):
    plt.figure()
    plt.plot(history.history['loss'], label='Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.legend()
    plt.grid()
    plt.title(title)
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.xlim([0, no_training_epochs - 1])
    plt.show()

# Simple Linear Model
model_linear = Sequential()
model_linear.add(Flatten(input_shape=(n_steps, 1)))
model_linear.add(Dense(1))

optimizer = Adam(learning_rate=0.001)
model_linear.compile(loss='mean_squared_error', optimizer=optimizer, metrics=['mean_squared_error'])
model_linear.summary()

history_linear = model_linear.fit(X_train, y_train, epochs=no_training_epochs, validation_data=(X_valid, y_valid))
plot_function(history_linear, "Linear Model Training History")

score_linear = model_linear.evaluate(X_test, y_test, verbose=0)
print('Linear Model Test Loss:', score_linear[0])
print('Linear Model Test MSE:', score_linear[1])

# Simple RNN Model
model_simple_rnn = Sequential()
model_simple_rnn.add(SimpleRNN(1, input_shape=[n_steps, 1]))

optimizer_simple_rnn = Adam(learning_rate=0.001)
model_simple_rnn.compile(loss='mean_squared_error', optimizer=optimizer_simple_rnn, metrics=['mean_squared_error'])
model_simple_rnn.summary()

history_simple_rnn = model_simple_rnn.fit(X_train, y_train, epochs=no_training_epochs, validation_data=(X_valid, y_valid))
plot_function(history_simple_rnn, "Simple RNN Training History")

score_simple_rnn = model_simple_rnn.evaluate(X_test, y_test, verbose=0)
print('Simple RNN Test Loss:', score_simple_rnn[0])
print('Simple RNN Test MSE:', score_simple_rnn[1])

# Deep RNN Model
model_deep_rnn = Sequential()
model_deep_rnn.add(SimpleRNN(20, return_sequences=True, input_shape=[n_steps, 1]))
model_deep_rnn.add(SimpleRNN(20, return_sequences=True))
model_deep_rnn.add(SimpleRNN(1))

optimizer_deep_rnn = Adam(learning_rate=0.001)
model_deep_rnn.compile(loss='mean_squared_error', optimizer=optimizer_deep_rnn, metrics=['mean_squared_error'])
model_deep_rnn.summary()

history_deep_rnn = model_deep_rnn.fit(X_train, y_train, epochs=no_training_epochs, validation_data=(X_valid, y_valid))
plot_function(history_deep_rnn, "Deep RNN Training History")

score_deep_rnn = model_deep_rnn.evaluate(X_test, y_test, verbose=0)
print('Deep RNN Test Loss:', score_deep_rnn[0])
print('Deep RNN Test MSE:', score_deep_rnn[1])
```

```python
##### Exercise 2 - The Movie Review

import keras
import numpy as np
import matplotlib.pyplot as plt
import time
from keras.models import Sequential
from keras.layers import SimpleRNN, LSTM, GRU, Dense, Embedding
from keras.preprocessing.sequence import pad_sequences
from keras.optimizers import Adam

# Loading Dataset
(X_train, y_train), (X_test, y_test) = keras.datasets.imdb.load_data(num_words=1000)

# Dataset Examination
word_index = keras.datasets.imdb.get_word_index()
id_to_word = {id_ + 3: word for word, id_ in word_index.items()}
for id_, token in enumerate(("<pad>", "<sos>", "<unk>")):
    id_to_word[id_] = token

single_sequence = " ".join([id_to_word.get(id_, "?") for id_ in X_train[0][:]])
print("Example review (decoded):", single_sequence)

# Preprocessing: Padding
max_length = 150
X_train = pad_sequences(X_train, maxlen=max_length, padding='post', truncating='post', value=0)
X_test = pad_sequences(X_test, maxlen=max_length, padding='post', truncating='post', value=0)

# Shared Configuration
embed_size = 128
vocabulary_size = 1000
learning_rate = 0.001
epochs = 5

# Defining training & evaluation function
def train_and_evaluate_model(model, name):
    model.compile(loss="binary_crossentropy", optimizer=Adam(learning_rate), metrics=["accuracy"])
    print(f"\nTraining {name} model...\n")
    start_time = time.time()
    history = model.fit(X_train, y_train, epochs=epochs, validation_split=0.2, verbose=2)
    elapsed = time.time() - start_time
    score = model.evaluate(X_test, y_test, verbose=0)
    print(f'{name} Test Loss: {score[0]:.4f}')
    print(f'{name} Test Accuracy: {score[1]*100:.2f}%')
    print(f'{name} Training Time: {elapsed:.2f} seconds')
    return history

# Simple RNN Model
model_rnn = Sequential([
    Embedding(vocabulary_size, embed_size, input_shape=[None]),
    SimpleRNN(128, return_sequences=True),
    SimpleRNN(128),
    Dense(1, activation='sigmoid')
])
history_rnn = train_and_evaluate_model(model_rnn, "Simple RNN")

# LSTM Model
model_lstm = Sequential([
    Embedding(vocabulary_size, embed_size, input_shape=[None]),
    LSTM(128, return_sequences=True),
    LSTM(128),
    Dense(1, activation='sigmoid')
])
history_lstm = train_and_evaluate_model(model_lstm, "LSTM")

# GRU Model
model_gru = Sequential([
    Embedding(vocabulary_size, embed_size, input_shape=[None]),
    GRU(128, return_sequences=True),
    GRU(128),
    Dense(1, activation='sigmoid')
])
history_gru = train_and_evaluate_model(model_gru, "GRU")

# Plot comparison
def plot_comparison(h_rnn, h_lstm, h_gru):
    plt.figure()
    plt.plot(h_rnn.history['val_accuracy'], label='Simple RNN Val Acc')
    plt.plot(h_lstm.history['val_accuracy'], label='LSTM Val Acc')
    plt.plot(h_gru.history['val_accuracy'], label='GRU Val Acc')
    plt.title("Validation Accuracy Comparison")
    plt.xlabel("Epoch")
    plt.ylabel("Accuracy")
    plt.legend()
    plt.grid()
    plt.show()

plot_comparison(history_rnn, history_lstm, history_gru)
```
